        <li>
                داده های پرت (Outlier) هارا به روش IQR حذف کنید
        </li>
        <li>
            <font color="red">نکته مهم:</font> استفاده از مقادیر پیشفرض Q1 و Q3 داده پیشنهاد نمی شود. این دیتاست داده های کمی دارد و حذف تعداد زیادی از داده ها توصیه نمی شود. شما می توانید از مقادیر کمتری/بیشتری برای Q1 و Q3 استفاده کنید. مثلا مقدار 0.20 به جای 0.25 برای Q1 و مقدار 0.80 برای Q3.(مقادیر را خودتان انتخاب کنید)
        </li>
    </font>
</div>
<br>

```python
# TODO: حذف داده های پرت
def remove_outliers_iqr(df):
    df_clean = df.copy()
    numeric_cols = df_clean.select_dtypes(include=['int64', 'float64']).columns

    for col in numeric_cols:
        q1 = df_clean[col].quantile(0.20)
        q3 = df_clean[col].quantile(0.80)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        df_clean = df_clean[(df_clean[col] >= lower_bound) & (df_clean[col] <= upper_bound)]

    return df_clean
```

```python
# TODO: Histogram رسم
def plot_numeric_histograms(df):
    numeric_cols = df.select_dtypes(include='number').columns
    n_cols = 3  # تعداد ستون‌ها در گرید نمودارها
    n_rows = (len(numeric_cols) + n_cols - 1) // n_cols  # تعداد ردیف‌ها

    plt.figure(figsize=(n_cols * 5, n_rows * 4))

    for i, col in enumerate(numeric_cols, 1):
        plt.subplot(n_rows, n_cols, i)
        sns.histplot(df[col], bins=30, kde=True, color='skyblue')
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()
```

```python
# TODO: GMM پیدا کردن تعداد کلاستر های مناسب

import matplotlib.pyplot as plt

n_components = range(1, 11)
aic_scores = []
bic_scores = []

for n in n_components:
    gmm = GaussianMixture(n_components=n, random_state=42)
    gmm.fit(df_scaled)
    aic_scores.append(gmm.aic(df_scaled))
    bic_scores.append(gmm.bic(df_scaled))

plt.figure(figsize=(10, 5))
plt.plot(n_components, aic_scores, label='AIC', marker='o')
plt.plot(n_components, bic_scores, label='BIC', marker='s')
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('AIC and BIC for GMM')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# TODO: توزیع خوشه ها با Scatterplot
cluster_means = df.groupby('cluster')[['Spent', 'Income']].mean()

# نام‌گذاری
labels_map = {}
spent_mean = cluster_means['Spent'].mean()
income_mean = cluster_means['Income'].mean()

for idx, row in cluster_means.iterrows():
    if row['Spent'] > spent_mean and row['Income'] > income_mean:
        labels_map[idx] = 'VIP'
    elif row['Spent'] < spent_mean and row['Income'] < income_mean:
        labels_map[idx] = 'Low Spender'
    else:
        labels_map[idx] = 'Average'

df['cluster_label'] = df['cluster'].map(labels_map)

plt.figure(figsize=(8,6))
sns.scatterplot(
    x='Spent', y='Income',
    hue='cluster_label',
    data=df,
    palette='Set2',
    alpha=0.7
)
plt.title('Customer Segmentation and Labeling')
plt.xlabel('Spent (Total Purchases)')
plt.ylabel('Income')
plt.legend(title='Customer Type')
plt.show()
```
